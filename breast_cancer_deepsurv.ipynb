```python
import pandas as pd

# Load your data
df = pd.read_csv('data/Recurrence free survival in breast cancer patients.csv')
```

```python
df
```

```python
# Assuming 'time' is the survival time column and 'event' indicates if the event occurred
covariates = df.drop(['time', 'status'], axis=1)
times = df['time']
events = df['status']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
covariates = scaler.fit_transform(covariates)
```

```python
from deepsurv import DeepSurv
from deepsurv.utils import make_surv_data

# Prepare the survival dataset for DeepSurv
surv_data = make_surv_data(covariates, times, events)

# Define the hyperparameters for the DeepSurv model
hyperparams = {
    'L2_reg': 0.001,
    'learning_rate': 0.01,
    'batch_norm': True,
    'dropout': 0.5
}

# Initialize the DeepSurv model
model = DeepSurv(hyperparams)

# Train the model
model.train(surv_data)
```

```python
from deepsurv.utils import concordance_index

# Predict the risk scores
risk_scores = model.predict_risk(covariates)

# Calculate concordance index
ci = concordance_index(risk_scores, times, events)
print(f"Concordance Index: {ci}")
```
