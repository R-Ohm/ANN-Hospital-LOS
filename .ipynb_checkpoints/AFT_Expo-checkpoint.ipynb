```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import ExponentialFitter
from lifelines.utils import concordance_index
import numpy as np
```

```python
# Load your dataset
data = pd.read_csv("data/Recurrence free survival in breast cancer patients.csv")  

# Fit the Exponential model
exp_fitter = ExponentialFitter()
exp_fitter.fit(durations=data['time'], event_observed=data['status'])

# Display the model summary
exp_fitter.print_summary()

lambda_param = 1 / exp_fitter.lambda_  # lambda parameter from the fitted model
cumulative_hazards = data['time'] / lambda_param  # cumulative hazard for each observation
```

```python
valid_idx = ~np.isnan(cumulative_hazards)
cleaned_times = data['time'][valid_idx]
cleaned_status = data['status'][valid_idx]
cleaned_hazards = cumulative_hazards[valid_idx]
```

```python
# Calculate the C-index for model evaluation
c_index = concordance_index(cleaned_times, -cleaned_hazards, cleaned_status)
print(f"C-index: {c_index}")
```

```python
# Plot the Survival Function
plt.figure(figsize=(10, 6))
exp_fitter.plot_survival_function()
plt.title('Survival Function (Exponential Model)')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.grid(True)
plt.show()
```

```python
# Plot the Cumulative Hazard Function
plt.figure(figsize=(10, 6))
exp_fitter.plot_cumulative_hazard()
plt.title('Cumulative Hazard Function (Exponential Model)')
plt.xlabel('Time')
plt.ylabel('Cumulative Hazard')
plt.grid(True)
plt.show()
```
